```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
%matplotlib widget

pi = np.pi
Lambda1 = 589e-9
Lambda2 = 589.6e-9
Largeur = 20e-12
c = 3e8
Frequence1 = c/Lambda1
Frequence2 = c/Lambda2
Periode1 = 1/Frequence1
Periode2 = 1/Frequence2
Omega1 = 2*pi*Frequence1
Omega2 = 2*pi*Frequence2

Nb_periodes = 100000
Echantillons = 32
Pas_temps = Periode1 / Echantillons
t = np.arange(-Nb_periodes/2*Periode1,Nb_periodes/2*Periode1,Pas_temps) # vecteur temps


Signal = np.zeros(len(t))

Signal = np.cos(Omega1*t) + np.cos(Omega2*t)
Signal = Signal * np.exp(-t*t/Largeur/Largeur)


plt.figure(figsize=(12,5))
plt.plot(t, Signal)



```

```python
Spectre = np.fft.fft(Signal);
Spectre = np.fft.fftshift(Spectre);
plt.figure(figsize=(12,5))

FreqResolution=Lambda1**2/c/(Pas_temps*Nb_periodes*Echantillons)
print(FreqResolution)

FreqDomaine=Lambda1**2/c/Pas_temps
print(FreqDomaine)

u = np.arange(-FreqDomaine/2,FreqDomaine/2,FreqResolution)
plt.plot(u,abs(Spectre))


```
